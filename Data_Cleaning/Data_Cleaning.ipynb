```python
import yfinance as yf
import pandas as pd
import numpy as np
import json
```

```python
ticker = yf.Ticker("005930.KS")
hist_data = ticker.history(period="max")
hist_data.to_csv("samsung_stock.csv")
```

```python
hist_data
```

```python
hist_data.index = hist_data.index.tz_localize(None)
```

```python
hist_data = hist_data.reset_index()
hist_data['Date'] = pd.to_datetime(hist_data['Date'])
```

```python
invalid_price = hist_data[
    (hist_data['Low'] > hist_data['High']) |
    (hist_data['Open'] > hist_data['High']) |
    (hist_data['Close'] > hist_data['High']) |
    (hist_data['Open'] < hist_data['Low']) |
    (hist_data['Close'] < hist_data['Low'])
]
```

```python
hist_data.shape
```

```python
# Chuyển DataFrame sang danh sách JSON (dạng records)
json_data = hist_data.to_json(orient='records', date_format='iso', indent=4)

# Lưu ra file
with open('hist_data_cleaned.json', 'w', encoding='utf-8') as f:
    f.write(json_data)
```

```python
# Chỉ lấy những ngày có sự kiện đặc biệt
events_json = hist_data[
    (hist_data['Dividends'] > 0) | (hist_data['Stock Splits'] > 0)
].to_json(
    orient='records',
    date_format='iso'
)
with open('samsung_special_events.json', 'w', encoding='utf-8') as f:
    f.write(events_json)
```

```python
import os

# 2. Set the target path (macOS local)
target_path = "/Users/nguyentoan/Documents/AI_Financial_RAG/Data" 
# Create folder if not exists
os.makedirs(target_path, exist_ok=True)

# 3. List of files to save
files_to_upload = {
    "hist_data_cleaned.json": json_data,
    "samsung_special_events.json": events_json
}

# 4. Save files
for file_name, content in files_to_upload.items():
    full_path = os.path.join(target_path, file_name)
    with open(full_path, "w", encoding="utf-8") as f:
        f.write(content)
    print(f"✅ Successfully saved: {full_path}")

print("\n--- Data preparation for AI Analyst completed ---")
```
